```python
from numpy import *
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("paper")
plt.figure(figsize=(6,4), dpi=200)
```

```python
r1 = 0.15
r2 = 0.20
r3 = 0.30 
R =linspace(0.01, 2,1000)
I1 = 1.5/(R+r1)
I2 = 1.5/(R+r2)
I3 = 1.5/(R+r3)
P1 = I1**2*R
P2 = I2**2*R
P3 = I3**2*R
plt.plot(R,P1, R,P2, R, P3)
plt.xlabel('R (Ohms)')
plt.ylabel('power (W)')
plt.show()
```

```python
E1 = P1/(P1 + I1**2*r1)
E2 = P2/(P2 + I1**2*r2)
E3 = P3/(P3 + I1**2*r3)
plt.plot(R,E1, R,E2, R, E3)
plt.xlabel('R (Ohms)')
plt.ylabel('fraction of power')
plt.show()
```
